# Modeling / Feature Engineering

## Model Trend

```python
# load train and test data 
train = pd.read_csv('data/train.csv',index_col=0,parse_dates=True)
test = pd.read_csv('data/test.csv',index_col=0,parse_dates=True)
```

```python
train.head()
```

```python
# engineer timestep variable for future convenience
train['timestep'] = list(range(train.shape[0]))
```

```python
X_trend = train[['timestep']]
y_trend = train['temp']
```

```python
# trend model

trend_model = make_pipeline(
    PolynomialFeatures(degree=2),
    LinearRegression()
)
```

```python
# fit model

trend_model.fit(X_trend,y_trend)
```

```python
train['trend'] = trend_model.predict(X_trend)
```

```python
mpl.rc('figure',figsize=(12,3),dpi=100)

train[['temp','trend']].resample('10Y').mean().plot(legend=True)
sns.despine()
```

```python
mpl.rc('figure',figsize=(12,3),dpi=100)

train[['temp','trend']].resample('3Y').mean().plot(legend=True)
sns.despine()
```

```python
# remove trend influence from data
train['detrended'] = train['temp'] - train['trend']
```

```python
mpl.rc('figure',figsize=(12,3),dpi=100)

train[['temp','detrended']].resample('10Y').mean().plot(legend=True)
sns.despine()
```

```python
mpl.rc('figure',figsize=(12,3),dpi=100)

train['detrended'].loc['1900':'1950'].plot(legend=True)
sns.despine()
```

## Model  Sesaonality

```python
# create month column
train['month'] = train.index.month
```

```python
# feature and target for seasonality model

X_seasonality = train[['month']]
y_seasonality = train['detrended']
```

```python
# seasonality model

seasonality_model = make_pipeline(
    OneHotEncoder(drop='first'),
    LinearRegression()
)
```

```python
# fit model

seasonality_model.fit(X_seasonality, y_seasonality)
```

```python
train['seasonality'] = seasonality_model.predict(X_seasonality)
```

```python
mpl.rc('figure',figsize=(12,3),dpi=100)

train[['detrended','seasonality']].loc['1950':'1960'].plot(legend=True)
sns.despine()
```

## Combine Trend & Seasonality

```python
# combine trend and seasonality

train['trend_seasonal'] = train['trend'] + train['seasonality']

mpl.rc('figure',figsize=(12,3),dpi=100)
train[['temp','trend_seasonal']].loc['1950':'1960'].plot(legend=True)
```

```python
train[['temp','trend_seasonal']].head()
```

```python
# leftover time series devoid of trend and seasonality

train['remainder'] = train['temp'] - train['trend_seasonal']
```

```python
mpl.rc('figure',figsize=(12,3),dpi=100)

train[['temp','remainder']].resample('10Y').mean().plot(legend=True)
sns.despine()
```

```python
# new loading of train and test
train = pd.read_csv('data/train.csv',index_col=0,parse_dates=True)
test = pd.read_csv('data/test.csv',index_col=0,parse_dates=True)
```

```python
# add timestep column
train['timestep'] = list(range(train.shape[0]))
```

```python
# add month column
train['month'] = train.index.month
```

```python
# timestep should continue after train 
test['timestep'] = list(range(train.shape[0],train.shape[0]+test.shape[0]))
```

```python
test['month'] = test.index.month
```

```python
# features and target for train set
Xb = train[['timestep','month']]
yb = train['temp']
```

```python
# features and target for test set
X_test_baseline = test[['timestep','month']]
y_test = test['temp']
```

```python
transformer = ColumnTransformer([
    ('poly', PolynomialFeatures(), ['timestep']),
    ('ohe', OneHotEncoder(drop='first'), ['month'])]
)
```

```python
estimator = Pipeline([
        ('transformer', transformer),
        ('lr', LinearRegression())]
)
```

```python
param_grid = {'transformer__poly__degree': [1,2,3,4]}
```

```python
cv = TimeSeriesSplit(n_splits=3)
```

```python
gscv = GridSearchCV(
    estimator=estimator,
    param_grid=param_grid,
    scoring='r2',
    cv=cv, 
    n_jobs=-1
)
```

```python
# initial time
ti = time.time()

# grid-search cross-validation
gscv.fit(Xb,yb)

# final time 
tf = time.time()

# time taken
print(f"time taken: {round(tf-ti,2)} sec")
```

```python
column_list = ['param_transformer__poly__degree',
               'mean_test_score',
               'std_test_score',
               'rank_test_score'
              ]
result_df = pd.DataFrame(gscv.cv_results_)[column_list]

result_df.rename(
    columns=lambda name: name.split('__')[-1],inplace=True
)
result_df
```
